1. Show me a breakdown of distance from home by job role and attrition.

```python
df.groupby(['DistanceFromHome', 'JobRole', 'Attrition'])['Attrition'].count().unstack().head(30)
```

```python
sns.catplot(x="JobRole", y="DistanceFromHome", hue="Attrition", data=df)
plt.xticks(rotation=45)
```

```python
df['DistanceFromHome'].value_counts()
```

```python
sns.countplot(x = 'DistanceFromHome', hue = 'Attrition', data = df)
plt.figure(figsize=(300,5))
```

```python
df['JobRole'].value_counts()
```

```python
sns.countplot(x = 'JobRole', hue = 'Attrition', data = df)
plt.xticks(rotation=45)
```
